```python
import os
import struct
import numpy as np

"""
Loosely inspired by http://abel.ee.ucla.edu/cvxopt/_downloads/mnist.py
which is GPL licensed.
"""

def read(dataset = "training", path = "."):
    """
    Python function for importing the MNIST data set.  It returns an iterator
    of 2-tuples with the first element being the label and the second element
    being a numpy.uint8 2D array of pixel data for the given image.
    """

    if dataset is "training":
        fname_img = os.path.join(path, '../Data/train-images.idx3-ubyte')
        fname_lbl = os.path.join(path, '../Data/train-labels.idx1-ubyte')
    elif dataset is "testing":
        fname_img = os.path.join(path, 't10k-images-idx3-ubyte')
        fname_lbl = os.path.join(path, 't10k-labels-idx1-ubyte')
    else:
        raise ValueError("dataset must be 'testing' or 'training'")

    # Load everything in some numpy arrays
    with open(fname_lbl, 'rb') as flbl:
        magic, num = struct.unpack(">II", flbl.read(8))
        lbl = np.fromfile(flbl, dtype=np.int8)

    with open(fname_img, 'rb') as fimg:
        magic, num, rows, cols = struct.unpack(">IIII", fimg.read(16))
        img = np.fromfile(fimg, dtype=np.uint8).reshape(len(lbl), rows, cols)

    get_img = lambda idx: (lbl[idx], img[idx])

    # Create an iterator which returns each image in turn
    for i in range(len(lbl)):
        yield get_img(i)

def show(image):
    """
    Render a given numpy.uint8 2D array of pixel data.
    """
    from matplotlib import pyplot
    import matplotlib as mpl
    fig = pyplot.figure()
    ax = fig.add_subplot(1,1,1)
    imgplot = ax.imshow(image, cmap=mpl.cm.Greys)
    imgplot.set_interpolation('nearest')
    ax.xaxis.set_ticks_position('top')
    ax.yaxis.set_ticks_position('left')
    pyplot.show()
```

```python
train_data = list(read(dataset='training',path='.'))
```

### Preparing the data
    In the cell below we are flatteing the data to a row(in data_all) and treating each row as a datapoint and the lables for each datapoint would be in label_all

```python
import numpy as np
data_all = []
label_all = []
for i in range(0,60000):
    data_all.append(train_data[i][1].flatten())
    label_all.append(train_data[i][0])
```

    Picking up only a few first points. On these points we would do dimensionality reduction by PCA

```python
data = data_all[0:sample_length]
data = np.array(data)
label = label_all[0:sample_length]
label = np.array(label)
data.shape, label.shape
```

Down in the cell we are Standardizing the data column wise

```python
import warnings; warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
stan_data = StandardScaler().fit_transform(data)
stan_data.shape, np.sum(stan_data)
```

## Using sklearn
    We would use sklearn library to do PCA on our chosen datapoints

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2,svd_solver='auto')
new_data = pca.fit_transform(stan_data)
new_data.shape
```

In the above cell we have applied PCA on our data and got back the coordinates of the axes in which our data is most explained.<br>
In the cell below we have plotted the data and also given color codes for the classes

```python
load_in_df1 = np.array([new_data[:,0],new_data[:,1],label])
load_in_df1 = load_in_df1.T
import pandas as pd
df_1 = pd.DataFrame(data=load_in_df1,columns=['PC1','PC2','class'])
dist_color = ["#e6194b", "#3cb44b", "#ffe119", "#0082c8", "#46f0f0",\
"#911eb4", '#aa6e28', '#000000','#f032e6','#aaffc3']
sns.FacetGrid(df_1, hue="class",aspect=2,palette=sns.color_palette(dist_color), size=6) \
   .map(plt.scatter, "PC1",'PC2') \
   .add_legend()
plt.show()
```

```python
VAR = pca.explained_variance_
print ('Variance in PC1: {}\nVariance in PC2: {}'.format(VAR[0],VAR[1]))
```

```python
VAR_ratio = pca.explained_variance_ratio_
print ('Variance-ratio in PC1: {}\nVariance-ratio in PC2: {}'.format(VAR_ratio[0],VAR_ratio[1]))
```

# Not using sklearn
    Self implemented on 2 axes
    
    Preparing the covariance matrix

```python
c_mat = np.matmul(stan_data.T, stan_data)
c_mat.shape
```

* The shape in 784x784 because we had 784 features and this c_mat contains the covariance between the features.
* C_mat is a hermitian matrix
* The diagonal elements of c_mat explain the variance of the ith feature

        In the cell below we are performing eigen-decomposition of the covariance matrix, upon doing so I'll   get the eigen vectors and eigen values of the covariance matrix

```python
import scipy as sc
val, vec = sc.linalg.eigh(c_mat)
val.shape, vec.shape
```

```python
val[-10:]
```

The eigen values are in the increasing order, thus the vectors corresponding to the eigen values(2 maximum) will also be present in the last position

```python
eig_vec = np.array([vec[:,-1],vec[:,-2]])# Arranging the eigen-vectors in an array
coord_matrix = np.matmul(eig_vec,stan_data.T)

load_in_df2 = np.array([coord_matrix[0,:],coord_matrix[1,:],label])
load_in_df2 = load_in_df2.T
df2 = pd.DataFrame(data=load_in_df2,columns=['PC1','PC2','class'])
sns.FacetGrid(df2, hue="class",aspect=2,palette=sns.color_palette(dist_color), size=6) \
   .map(plt.scatter, "PC1",'PC2') \
   .add_legend()
plt.show()
```

```python
plt.plot(val[::-1])
plt.show()
```

```python
print ('Variance-ratio in PC1: {}\nVariance-ratio in PC2: {}'.format(val[-1]/val.sum(),val[-2]/val.sum()))
```

We can see that the variance ratio we get while implementin PCA manually is the same while using sklearn library.<br>Therefore we can conclude that our PCA manual implementation was correct.
